```python
import numpy as np
from matplotlib import pyplot
```

```python
data = np.loadtxt('data1.txt', delimiter=',')
X, y = data[:, 0],data[:, 1]
m = y.size
```

```python
def plotData(x,y):
    fig =pyplot.figure()
    pyplot.plot(x, y, 'ro', ms=10, mec='k')
    pyplot.ylabel('Profit')
    pyplot.xlabel('Population')

plotData(X,y)
```

```python
# h(x) = theta0 + theta1.x

# theta0 = 0
# theta1 = 0
# theta = [0, 0]

# h(10) = 0 + 0.10 = 0
# h(20) = 0 + 0.20 = 0
# h(30) = 0 + 0.30 = 0
# h(40) = 0 + 0.40 = 0

# X = [10, 20, 30, 40]
# theta = [0, 0]


# A = [[1, 10], [1, 20], [1, 30]].[1, 1]
#   = [11, 21, 31]

# A = [[1,2,3,4], [1,2,3,4]]
# B = [2,3,4,5]
# A.B = [40, 40]

X = np.stack([np.ones(m), X], axis=1)
```

```python
def computeCost(X, y, theta):
    m = y.size
    h = np.dot(X, theta)

    J = (1/(2*m))*np.sum(np.square(h-y))

    return J

X
```

```python
J = computeCost(X, y, theta=np.array([1, 2]))
print('With theta = [1, 2] \nCost = %.2f' % J)

J = computeCost(X, y, theta=np.array([0, 0]))
print('With theta = [0, 0] \nCost = %.2f' % J)
```

```python
def gradientDescent(X, y, theta, alpha, num_iteration):
    m = y.size
    
    theta = theta.copy()
    J_history = [] 

    for i in range(num_iteration):
        theta = theta - (alpha/m) * (np.dot(X, theta) - y).dot(X)
        J_history.append([i, computeCost(X, y, theta)])

    return theta, J_history
```

```python
theta = np.zeros(2)

iterations = 1000
alpha = 0.002

theta, J_history = gradientDescent(X, y, theta, alpha, iterations)
J_history
```

```python
plotData(X[:, 1], y)
pyplot.plot(X[:, 1], np.dot(X, theta), '-')
pyplot.legend(['Training data', 'Linear regression'])

J_history = np.array(J_history)
fig = pyplot.figure()
pyplot.plot(J_history[:, 0], J_history[:, 1], '-')
pyplot.ylabel('J(theta)')
pyplot.xlabel('no of iterations')
# pyplot.axis([0, 5, 0, 100])
```
